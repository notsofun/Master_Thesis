```python
import pandas as pd
import os

# 修改为你的实际路径
csv_path = "chinese_finetuning_1_with_split.csv" 
label_col = "hate_speech" # 你的标签列名

def inspect_csv(path, col_name):
    if not os.path.exists(path):
        print(f"找不到文件: {path}")
        return

    df = pd.read_csv(path)
    
    print(f"--- 检查列: {col_name} ---")
    print(f"总行数: {len(df)}")
    
    # 1. 统计空值
    nan_count = df[col_name].isna().sum()
    print(f"NaN (空值) 数量: {nan_count}")
    
    # 2. 列出所有唯一值（看看有没有脏数据，比如空格、回车等）
    unique_values = df[col_name].unique()
    print(f"列中的所有唯一取值: {unique_values}")
    
    # 3. 打印出具体是哪几行报错 (NaN)
    if nan_count > 0:
        print(f"\n[警告] 以下索引行存在 NaN:")
        print(df[df[col_name].isna()].index.tolist()[:20], "... (仅展示前20条)")

# 使用示例
inspect_csv(csv_path, label_col)
```
