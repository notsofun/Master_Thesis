```python
import pandas as pd

# 1. 定义你的关键词列表
keywords = [
    "上帝", "耶稣", "基督", "基督徒", "教会", "圣经", "圣灵", "圣餐", "圣体",
    "十字架", "旧约", "新约", "福音", "教义", "做礼拜", "宗教", "教区", 
    "教派", "天主", "天主教", "东正教", "新教", "犹太教", "以色列", 
    "传教士", "神父", "牧师", "主教", "教皇", "修道士", "修女", "修道院",
    "耶狗", "基督狗", "神棍", "布道鬼", "布道佬", "皮条教", "圣母婊",
    "脑残信徒", "二毛子", "性罪教", "十字军党", "圣战", "十字军", 
    "东征", "审判", "教义争论", "改革", "宗教改革", "中世纪"
]

# 2. 读取数据（建议先读取一部分测试，或者全量读取）
df = pd.read_csv('extract_zh_religious.csv', usecols=['text'])

# 3. 核心检测函数：返回命中的关键词（如果没有命中返回 None）
def find_keywords(text):
    if pd.isna(text):
        return None
    # 找出所有包含在文本中的关键词
    hits = [word for word in keywords if word in str(text)]
    return ",".join(hits) if hits else None

# 4. 执行筛选
# 创建新列记录命中的词，方便你核对
df['hit_words'] = df['text'].apply(find_keywords)

# 过滤掉没有命中的行
filtered_df = df[df['hit_words'].notna()].copy()

# 5. 查看结果
print(f"原始数据量: {len(df)}")
print(f"命中关键词的数据量: {len(filtered_df)}")

# 预览前10条结果，重点看文本和命中的词
pd.set_option('display.max_colwidth', 500) # 让文本列显示得更长一点
display(filtered_df[['hit_words', 'text']].head(10))
```
